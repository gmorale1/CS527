```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
"""
graphs simulation results.

y: fitness, x: generation
defaults: pop size = 50, mutation: 0.01, crossover: 0.3, tournament: 2

results columns: iteration num,generation,avg fitness,best fitness,best genome,solved,num solutions,diversity,n,p_m,p_c,trn_size
"""
results = pd.read_csv("results.csv")

#graph populations
populations = []
populations.append(results[(results["n"] == 25 ) & (results["p_m"] == 0.01) & (results["p_c"] == 0.3) & (results["trn_size"] == 2)])
populations.append(results[(results["n"] == 50 ) & (results["p_m"] == 0.01) & (results["p_c"] == 0.3) & (results["trn_size"] == 2)])
populations.append(results[(results["n"] == 75 ) & (results["p_m"] == 0.01) & (results["p_c"] == 0.3) & (results["trn_size"] == 2)])
populations.append(results[(results["n"] == 100 ) & (results["p_m"] == 0.01) & (results["p_c"] == 0.3) & (results["trn_size"] == 2)])
for population in populations:
    x = []
    y = []

    pop_data = pd.DataFrame({
        "generation":[],
        "avg fitness":[],
        "best fitness":[],
    })

    for iteration in population["iteration num"]:
        x.append(population[population["iteration num"] == iteration]["generation"].tolist())
        y.append(population[population["iteration num"] == iteration]["avg fitness"].tolist())

        gen = population[population["iteration num"] == iteration]["generation"].tolist()
        avg = population[population["iteration num"] == iteration]["avg fitness"].tolist()
        best = population[population["iteration num"] == iteration]["best fitness"].tolist()

        iter_data = pd.DataFrame({
            "generation":gen,
            "avg fitness":avg,
            "best fitness":best
        })

        pop_data = pd.concat([pop_data,iter_data],ignore_index=True)

    #plot population
    #graphs[0].plot(x,y)
    sns.lineplot(data = pop_data, x = "generation", y = "avg fitness")
    sns.lineplot(data = pop_data, x = "generation", y = "best fitness")
plt.show()


```
